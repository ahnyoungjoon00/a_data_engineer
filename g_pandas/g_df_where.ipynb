```python
emp = pd.read_csv("./data/emp.csv")
emp.head()
```

```python
# 성이 '이'씨인 사원의 이름과 연봉 조회
emp["SALARY_2"] = emp["SALARY"]*12
emp[emp["EMP_NAME"].str.startswith("이")][["EMP_NAME", "SALARY_2"]]

# 부서코드가 D9인 사원의 이름과 부서코드
emp[emp["DEPT_CODE"] == "D9"][["EMP_NAME", "DEPT_CODE"]]

# 성별이 여성인 사원을 모두 조회, 2000년생 이후 사원은 없음
emp[["BIRTH", "SECRET"]] = emp["EMP_NO"].str.split('-', expand=True)
emp[(emp["SECRET"].str.startswith("2"))]

emp.loc[emp["EMP_NO"].str.find("2", 7, 8) != -1]

# 입사연도 1999년 이후인 사원의 이름과 입사일자를 조회
emp["HIRE_DATE"] = pd.to_datetime(emp["HIRE_DATE"])
emp[emp["HIRE_DATE"].dt.year > 1999]
```
