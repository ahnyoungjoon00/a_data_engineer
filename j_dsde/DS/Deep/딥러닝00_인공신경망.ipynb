## 로지스틱 회귀로 패션 아이템 분류하기

```python
# 이미지 전처리
# 흑백 이미지이므로 0-255의 값을 갖는 2차원 : 255로 나눠서 숫자 범위를 줄일예정
train_scaled = train_input/255

print(train_scaled.shape) 
# 이미지 한장을 1차원으로 변환 : 한장의 이미지 픽셀수는 29*28 = 784
train_scaled = train_scaled.reshape(-1, 28*28)
print(train_scaled.shape) # 784픽셀
```

#### SGDClassifier 알고리즘
- 확률적 경사 하강법(Stochastic Gradient Decent:SGD)을 통해 logistic regression 등의 선형 분류기를 학습시킬 수 있는 분류기
- 확률적 경사 하강법
    - 각 학습마다 모든 배치(sample)가 아닌 일부 샘플을 랜덤으로 선택하여 학습에 이용한다는 것이 '확률적'의 의미
    - 랜덤으로 선택된 n개의 배치를 학습에 이용하여 그 평균 경사를 구함
- 경사하강법 적용 시 사용되는 샘플배치가 달라지는것을 제외하면 logistic regression과 동일함

```python
# !pip3 install scikit-learn
```

```python
from sklearn.model_selection import cross_validate
from sklearn.linear_model import SGDClassifier
import numpy as np
# max_iter=5 반복횟수
sc = SGDClassifier(max_iter=5, random_state=42, loss="log_loss")
scores = cross_validate(sc, train_scaled, train_target, n_jobs=-1) # 학습 및 검증

print(scores)
print("-------------------")
print(np.mean(scores['test_score'])) # 검증점수 확인

# 학습데이터 정확도는 81.9% 정도 나왔음
# {'fit_time': array([8.13843656, 8.11638594, 8.14466763, 8.24287271, 8.17099524]), 
#  'score_time': array([0.0683372 , 0.0686636 , 0.06761813, 0.06086373, 0.06135702]), 
#  'test_score': array([0.83      , 0.81991667, 0.82008333, 0.80766667, 0.81958333])}

# 0.81945
```

### 로지스틱 회귀 식
- 회귀 알고리즘이므로 회귀계수를 구해 각 결정값을 만들어낸다
- z_티셔츠 = w1 * 픽셀1 + w2 * 픽셀2 + ..... + w784 * 픽셀784 + b
    - 10개의 분류이므로 10개의 방정식이 생성되고 10개 방정식에 대해 모든 데이터를 적용하여 각 결정값의 확률을 찾아냄
    

## 인공신경망
- 확률적 경사하강법을 사용하는 로지스틱 회귀와 같다
- z값(결정값)을 계산하는 단위를 뉴런이라고 함(요즘은 유닛이라고 부름)
    - 이 값을 바탕으로 클래스를 예측하기때문에 출력층 이라고도 부름
- 각 픽셀을 입력층이라고 부름
    - 입력층은 픽셀 자체이고 특별한 계산을 수행하지는 않는다
- 출력층 Zi를 만들기 위해서 픽셀 1(x1으로 표현, 머신러닝의 feature로 생각하면 됨)에 곱해지는 가중치는 W1.1, 픽셀 2(X2)에 곱해지는 가증치는 W1.2 라고 표기한다
- 절편은 뉴런(결정값)마다 하나씩임 (화면 그림에서는 b로 표현됨) 



- 인공 뉴런을 그림으로 표현하면 모양이 생물학적 뉴런과 비슷하여 부르는 이름이지 생물학적 뉴런의 기능과 아무 상관 없음
    - 새로운 종류의 머신러닝 알고리즘일 뿐임

## 인공신경망으로 모델 만들기

- 머신 러닝은 교차검증을 사용해 모델을 평가했지만 딥러닝은 검증 데이터를 따로 덜어내어 사용함
    - 딥러닝의 데이터는 충분히 크기때문에 검증 검수가 안정적임
    - 교차 검증을 수행하기에는 시간이 너무 오래 걸리기 때문에 딥러닝에서는 활용하지 않는다

```python
from sklearn.model_selection import train_test_split
# mnist는 충분한 크기는 아니지만 관례대로 검증데이터를 따로 써서 진행해볼 것임
train_scaled, val_scaled, train_target, val_target = train_test_split(
    train_scaled, train_target, test_size=0.2, random_state=42)
```

```python
print(train_scaled.shape, train_target.shape)
```

- 인공신경망 오른쪽을 생성(첫번째 층)
    - 10개의 패션 아이템을 구별하기 위해 10개의 뉴런(unit)으로 구성
    - softmax 함수(숫자를 0~1사이의 값으로 변경해 주는)를 통해 각 아이템의 비율을 계산
    - 총 몇개의 입력이 들어오는지 정보를 넣어줘야 함

- 밀집층
    - 케라스에는 다양한 층이 준비되어 있음
    - 밀집층은 784 * 10으로 연결되어 빽빽한 연결선이 있어 밀집층이라고 함

## 케라스 모델 구성
- 밀집층 생성을 위한 기본 모델 모양 구성
    - Dense 클래스 사용
    - Dense(출력층 수,activation=확률활성화함수, input_shape= 입력크기)
- 뉴런갯수, 뉴런 출력에 적용할 활성화함수, 입력의 크기를 파라미터로 사용함
- 10개의 출력값을 확률값으로 바꾸기 위해 activation=softmax 함수를 사용
    - 다중분류 : softmax , 이진분류 : sigmoid
- 밀집층 생성
    - Sequential 클래스 이용 : 구성정보를 담고 있는 dense 클래스 객체를 파라미터로 전달
    - model = Sequential(모델기본모양객체:dense)
    - model이 신명망모델 객체임 : 이 객체 인스턴스를 이용해 작업

```python
# 밀집층 생성을 위한 기본 모델 모형 구성
# 첫층 : 회귀 계수 찾는 연산 집행 전에 데이터를 주입하는 역할(input_dim 파라미터 사용) : 입력층 생성
# 모든층(Dense): 첫번째 파라미터는 뉴런(unit)수 >> 출력 수 >> 출력층 생성 >> 출력층은 Dense가 1개면 마지막 출력 >> 예측값 계산
# Dense가 여러개면 최종 Dense에서 예측값을 계산한다

# activation= 파라미터는 최종 예측과 연결된 Dense에서는 이진분류/멀티분류냐에 따라 달라짐
# activation= 파라미터는 중간층 Dense 면 비선형함수 Relu 또는 tanh 함수를 사용


input_dim = train_scaled.shape[1] # 위에서는 784 길이의 배열을 받는거였다면
output_layer = len(set(train_target))
# Dense에서 구조를 설정해서 이 구조대로 입력을 받겠다.
dense = keras.layers.Dense(output_layer, activation="softmax", input_dim = input_dim)
                   # 10개의 뉴런을 쓸거고, 
                   # 확률활성함수는 softmax 쓸거고, 
                   # 한번에 784개의 입력값을 받는다
```

```python
# Dense 클래스로 주입받아서 모델(밀집층) 생성
model = keras.Sequential()
model.add(dense)
```

- 위 코드로 인해 화면 그림과 같은 신경망 모델이 만들어짐

## 인공신경망으로 패션 아이템 분류하기
- 훈련전에 설정 단계가 있음
    - MODEL.compile(loss=손실함수,metrics=평가방법)
    - 손실함수란? 어떤 문제에 대해 머신러닝(딥러닝)이 얼마나 잘못되고 있는지를 측정하는 기준
        - 값이 작을수록 좋지만 최소값인지는 알지 못함
        - 가능한 많이 찾아보고 만족할만한 수준이라면 최소값이라고 인정하고 종료해야 함
        - 대부분 문제에서 다루는 손실함수들은 이미 정의되어 있는 경우가 많음
        - 이진분류 : binary_crossentropy
        - 다중분류 : categorical_crossentropy
    - 손실함수는 정답(y(타깃값))와 예측(^y)를 입력으로 받아 오차 점수를 만드는데, 이 점수가 높을수록 모델이 안좋은 것

**티셔츠 샘플에 대하여**
- 이진분류 형태처럼 계산(크로스엔트로피)
- 
- 
- 출력층은 10개의 뉴런이 있고 10개의 클래스에 대한 확률을 출력
    - 첫번째 출력 : 티셔츠일 확률, 두번째 출력 : 바지일 확률 ...
    - 타깃에 해당하는 확률만 남겨놓기 위해서는 나머지 확률에는 모두 0을 곱해줘야 함
    - 신경망은 티셔츠 샘플에서 손실을 낮추려면 첫번째 뉴런의 활성화 출력 a1이 값을 가능한 1에 가깝게 만들어야 함

- 샘플이 바지일 확률이 0.3이고 이 샘플의 분류값(정답)은 티셔츠라면
- \-log(0.3)*0 이 계산되므로 사라짐
- 즉 **손실함수는 각 타겟에 대하여 모델이 예측한 확률값을 받아서 해당 타겟의 예측에 대한 손실값만 남기고 나머지는 제거해야 함**
    - 그래서 손실함수 계산 시 타겟 데이터의 형태는 one-hot encoding 형태여야 한다


**원-핫 인코딩 형태**
- 크로스 엔트로피를 다중분류에서 쓰려면 원-핫 인코딩형태로 변형해야 함


- mnist의 타겟값은 0~9 까지의 정수
    - 원-핫 인코딩개념을 적용하여 손실값을 계산해주는 계산방법이   sparse_categorical_crossentropy 임

```python
model.compile(loss="sparse_categorical_crossentropy", metrics=["accuracy"])
```
