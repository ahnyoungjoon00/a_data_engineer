```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import pandas as pd
credit_df = pd.read_csv('./data/credit_final.csv')
credit_df.head()
```

```python
labels = credit_df['credit.rating']

# 피처 데이터 세트 
feature_data = credit_df.drop('credit.rating', axis=1)

feature_data.head(2)
```

```python
credit_df.info()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(feature_data,
                                                    labels,
                                                    test_size=0.2, 
                                                    random_state=11)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

# 의사 결정트리 객체 생성
dt_clf = DecisionTreeClassifier()
dt_clf.fit(X_train, y_train)
```

```python
# (1) DecisionTreeClassifier 학습 후 예측 수행
y_pred = dt_clf.predict(X_test)
print(f'DecisionTreeClassifier 정확도 : {accuracy_score(y_test, y_pred):.4f}')
```

```python
from sklearn.model_selection import cross_val_score

# cross_val_score() 함수 호출하고 5회 교차 검증 결과 정확도 반환 

scores = cross_val_score(dt_clf, feature_data, labels, cv=5)

for count, accuracy in enumerate(scores):
    print(f'교차검증 {count+1} 정확도 : { accuracy:.4f}')

print(f'평균 정확도 : {np.mean(scores):.4f}')
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
# KFold 활용 5회 교차 검증 결과 정확도 반환 
# random_state 42지정
kfold = KFold(n_splits=5, shuffle=True, random_state=42)
scores = cross_val_score(dt_clf, feature_data, labels, cv=kfold)

for count, accuracy in enumerate(scores):
    print(f'교차검증 {count+1} 정확도 : { accuracy:.4f}')

print(f'평균 정확도 : {np.mean(scores):.4f}')
```

```python
#성능평가지표 출력
from sklearn.metrics import confusion_matrix, accuracy_score, roc_auc_score, precision_score, recall_score, f1_score

# 실제값, 예측값, 예측 확률
def get_clf_eval(y_test, y_pred, pred_proba):
    cm = confusion_matrix(y_test, y_pred) # 혼동행렬 (실제값, 예측값)
    accuracy = accuracy_score(y_test, y_pred) # 정확도
    precision = precision_score(y_test, y_pred) # 정밀도
    recall = recall_score(y_test, y_pred)  #재현율
    f1 = f1_score(y_test, y_pred)   # f1 score
    roc_auc = roc_auc_score(y_test, pred_proba) # ROC_AUC score

    #오차행렬
    print('오차행렬 : ',cm)
    print(f'정확도 : {accuracy:.4f}, 정밀도 : {precision:.4f}, 재현율 : {recall:.4f}, F1 score : {f1:.4f}, ROC_AUC : {roc_auc:.4f}')
    
```
