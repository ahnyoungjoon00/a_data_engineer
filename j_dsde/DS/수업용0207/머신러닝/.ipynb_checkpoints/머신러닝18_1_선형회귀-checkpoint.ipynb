# 선형 회귀

## k-최근접 이웃의 한계
- 길이가 매우 긴 생선이 들어왔을 때 무게 예측?

```python
import numpy as np

perch_length = np.array(
    [8.4, 13.7, 15.0, 16.2, 17.4, 18.0, 18.7, 19.0, 19.6, 20.0, 
     21.0, 21.0, 21.0, 21.3, 22.0, 22.0, 22.0, 22.0, 22.0, 22.5, 
     22.5, 22.7, 23.0, 23.5, 24.0, 24.0, 24.6, 25.0, 25.6, 26.5, 
     27.3, 27.5, 27.5, 27.5, 28.0, 28.7, 30.0, 32.8, 34.5, 35.0, 
     36.5, 36.0, 37.0, 37.0, 39.0, 39.0, 39.0, 40.0, 40.0, 40.0, 
     40.0, 42.0, 43.0, 43.0, 43.5, 44.0]
     )
perch_weight = np.array(
    [5.9, 32.0, 40.0, 51.5, 70.0, 100.0, 78.0, 80.0, 85.0, 85.0, 
     110.0, 115.0, 125.0, 130.0, 120.0, 120.0, 130.0, 135.0, 110.0, 
     130.0, 150.0, 145.0, 150.0, 170.0, 225.0, 145.0, 188.0, 180.0, 
     197.0, 218.0, 300.0, 260.0, 265.0, 250.0, 250.0, 300.0, 320.0, 
     514.0, 556.0, 840.0, 685.0, 700.0, 700.0, 690.0, 900.0, 650.0, 
     820.0, 850.0, 900.0, 1015.0, 820.0, 1100.0, 1000.0, 1100.0, 
     1000.0, 1000.0]
     )
```

```python
from sklearn.model_selection import train_test_split

train_input, test_input, train_target, test_target = train_test_split(
    perch_length, perch_weight, random_state=42)

train_input = train_input.reshape(-1, 1)
test_input = test_input.reshape(-1, 1)
```

```python
from sklearn.neighbors import KNeighborsRegressor

knr = KNeighborsRegressor(n_neighbors=3)

knr.fit(train_input, train_target)
```

- 길이가 50cm인 농어의 무게를 예측

```python
print(knr.predict([[50]]))
# print(knr.predict([[42]]))
```

- 농어의 길이가 50cm면 무게는?
    - 1033 ?? 적당한 무게인지

```python
import matplotlib.pyplot as plt
```

```python

distances, indexes = knr.kneighbors([[50]])

plt.scatter(train_input, train_target)

plt.scatter(train_input[indexes], train_target[indexes], marker='D')

plt.scatter(50, 1033, marker='^')
# plt.scatter(42, 1066, marker='^')
plt.show()
```

```python
print(np.mean(train_target[indexes]))
```

```python
print(knr.predict([[100]]))
```

```python

distances, indexes = knr.kneighbors([[100]])


plt.scatter(train_input, train_target)

plt.scatter(train_input[indexes], train_target[indexes], marker='D')

plt.scatter(100, 1033, marker='^')
plt.show()
```

## 선형 회귀
- 널리 사용되는 대표적인 회귀 알고리즘
    - 비교적 간단하고 성능이 뛰어남
    - 특성이 하나인 경우 직선을 학습하는 알고리즘
        - 특성을 잘 나타내는 직선을 찾아야 한다.

## 다음 아래 3개의 이미지에서 그래프를 가장 잘 나타내는 직선은?




- from sklearn.linear_model import LinearRegression
    - 위 클래스로 구현되어 있음

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(train_input, train_target)
```

```python
lr.predict([[50]])
```

- K-최근접 이웃 알고리즘이랑은 다르게 50CM의 생선의 무게를 더 크게 예측함
- 어떻게 이 값이 나왔는지 직선으로 확인
- 직선을 그리기 위해서는 기울기와 절편이 필요함
    - 선형 회귀알고리즘은 모델객체의 속성으로 기울기와 절편을 제공함
        - coef_ : 기울기(계수, 가중치 등으로 불림)
        - intercept_ : 절편
 

```python
print(lr.coef_, lr.intercept_)
```

- 생선 길이 15에서 50까지의 무게를 예측해서 직선을 그린다고 하면
    - 직선의 두 점은
        - (15,15X39-709) , (50,50X39-709)

```python

plt.scatter(train_input, train_target)
# 15~50 사이의 길이에 대한 1차 방정식 직선 표현
plt.plot([15, 50], [15*lr.coef_+lr.intercept_, 50*lr.coef_+lr.intercept_])

plt.scatter(50, 1241.8, marker='^')
plt.show()
```

- 위 그래프를 보면 무게는 1200 정도로 예측 됨

```python
# 훈련세트와 테스트세트에 대한 R2 스코어 확인
print(lr.score(train_input, train_target))
print(lr.score(test_input, test_target))
```

- 훈련세트에 과대적합 되어 있는건지?
    - 그래프 왼쪽 아래를 보면 15cm의 무게는 -로 계산하고 있음
    - **길이와 무게의 산점도는 일직선이라기보다 왼쪽 위로 조금 구부러진 곡선에 가까움**

### 최적의 곡선 찾기
- 이차 방정식으로 그래프를 그린다
    - 길이를 제곱한 항이 훈련세트에 추가되어야 함
- 현재 특성은 길이 정보만 있음
    - 길이 특성 하나는 1차항으로 표현
    - 곡선을 찾기 위해 인위적으로 2차항을 추가 : 길이 특성을 제곱한 데이터를 추가

## 다항 회귀

```python
# 훈련세트의 길이 데이터를 2차항으로 데이터 추가 :열추가(column_stack()함수 사용)
train_poly = np.column_stack((train_input**2, train_input))
test_poly = np.column_stack((test_input**2, test_input))
```

```python
print(train_poly.shape, test_poly.shape)
```

- 제곱열을 추가했기 때문에 열이 늘어남

```python
train_poly[:10,:]
```

- 특성(제곱열)이 추가되었으므로 다시 학습

```python
lr = LinearRegression()
lr.fit(train_poly, train_target)
# 예측할때, 길이가 50인 생선의 무게 예측
print(lr.predict([[50**2, 50]]))
```

```python
print(lr.coef_, lr.intercept_)
```

### 아래와 같은 그래프를 학습함
- **회귀식** : 무게 = 1.0143321 * 길이**2 -21.55792498 * 길이 + 116.0502107827827

```python
train_poly[:, 1]
```

```python

point = np.arange(15, 50)
plt.scatter(train_poly[:, 1], train_target)
plt.scatter(50, 1573.98)
plt.plot(point, 1.0143321 * point**2 -21.55792498 * point + 116.0502107827827)
plt.show()
```

- 단순 선형 회귀모델보다 데이터에 적합한 그래프가 그려짐

```python
print(lr.score(train_poly, train_target))
print(lr.score(test_poly, test_target))
```

- 훈련세트와 테스트 점수가 높아짐


- 회귀 모델은 학습데이터의 범위를 벗어나는 범위의 데이터도 예측이 가능
