### (2) 데이터 세트 분리 : 학습 데이터 / 테스트 데이터 세트 
- 테스트 데이터 세트 : 30% 

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(cancer_data,
                                                    cancer_label,
                                                    test_size=0.3,
                                                    random_state=11)
```

### (3) 모델 생성 / 학습 / 예측 수행  
- 모델 생성 : 의사결정트리 알고리즘 사용   
- 학습 : 데이터 세트 기반으로 ML 알고리즘을 적용하여 모델 학습  
- 예측 수행 : 학습된 ML 모델을 이용해서, 테스트 데이터의 분류 예측 

```python
# 모델 생성 
# DecisionTreeClassifier 객체 생성
from sklearn.tree import DecisionTreeClassifier

dt_clf = DecisionTreeClassifier(random_state=11)
# 학습 : fit(학습용 피처 데이터, 학습용 레이블 데이터(정답))
# 학습용 데이터 세트로 학습 수행
dt_clf.fit(X_train, y_train)
```

```python
#예측 수행 : 학습된 ML 모델을 이용해서, 테스트 데이터의 분류 예측 
y_pred = dt_clf.predict(X_test)
# 학습이 완료된 DecisionTreeClassifier 객체에서 테스트 데이트 세트로 예측 수행
# prdict(테스트용 피처 데이터)
# 예측된 데이터 세트 반환
y_pred
```

    - 예측 정확도

### breast_cancer 데이터 세트로 데이터프레임 생성

```python
# 데이터 로드
cancer = load_breast_cancer()
```

```python
# 데이터프레임 생성
cancer_df = pd.DataFrame(data=cancer.data, columns=cancer.feature_names)
cancer_df['target'] = cancer.target
cancer_df
```

```python
cancer_data = cancer.data
cancer_data.shape # (569, 30)

cancer_label = cancer.target
cancer_label.shape # (569,)
```

```python
#####################################################################################
```

### 교차 검증 수행

```python
# (1) cross_val_score()
from sklearn.model_selection import cross_val_score
import numpy as np

scores = cross_val_score(dt_clf, cancer_data, cancer_label, cv=5)
#scores

for count, accuracy in enumerate(scores):
    print(f"교차검증 {count+1} 정확도 : {accuracy:.4f}")

print(f"평균 정확도 : {np.mean(scores):.4f}")
```

```python
# (2) KFold 사용
from sklearn.model_selection import KFold

kfold = KFold(n_splits=5, shuffle=True, random_state=42)
scores = cross_val_score(dt_clf, cancer_data, cancer_label, cv=kfold)


for count, accuracy in enumerate(scores):
    print(f"교차검증 {count+1} 정확도 : {accuracy:.4f}")

print(f"평균 정확도 : {np.mean(scores):.4f}")
```

### 하이퍼 파라미터 튜닝 수행

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV

dt_clf = DecisionTreeClassifier(random_state=11)
dt_clf.fit(X_train, y_train)
# 하이퍼 파라미터 설정 : 각 값을 조합하여 수행  
parameters = {
    "max_depth":[3, 5],
    "min_samples_leaf":[1, 6],
    "min_samples_split":[1, 2, 3]
}

# GridSearchCV(모델, 파라미터, 평가방법, 폴드 수)
grid_dclf = GridSearchCV(cv=5, estimator=dt_clf,
             param_grid=parameters,
             scoring='accuracy')
grid_dclf.fit(X_train, y_train)

print(f"GridSearchCV 최적의 하이퍼 파라미터 : {grid_dclf.best_params_}")
print(f"GridSearchCV 최고 정확도 : {grid_dclf.best_score_ :.6f}")

# GridSearchCV 최적의 하이퍼 파라미터 : {'max_depth': 3, 'min_samples_leaf': 6, 'min_samples_split': 2}
# GridSearchCV 최고 정확도 : 0.934525
```

```python
# 최고 점수를 낸 파라미터를 가진 모형(모델)
best_dclf = grid_dclf.best_estimator_

# best_dclf (최적 하이퍼 파라미터로 학습된 estimator(모델)로 예측/평가 수행)
y_pred = best_dclf.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)

print(f"하이퍼 파라미터 튜닝을 끝낸 최종 정확도 : {accuracy:.4f}")

y_pred = dt_clf.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)

print(f"하이퍼 파라미터 튜닝 전 예측 정확도 : {accuracy:.4f}")
# 하이퍼 파라미터 튜닝을 끝낸 최종 정확도 : 0.9415
# 하이퍼 파라미터 튜닝 전 예측 정확도 : 0.9474
```
