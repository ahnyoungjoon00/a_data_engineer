# 회귀 (Regression)

### 회귀 (Regression)
- 분류와 더불어 지도학습의 가장 큰 축 중의 하나
- 지도학습에서 가장 많이 활용되는 분야  
- 사실 분류보다 회귀가 현업 업무 분석가들이나 데이터 분석가들이 많이 활용
    - 대부분 기업에서 사용하는 예측이 수치적 예측이 많기 때문 (다음 달 매출액이 얼마인지 예측)  


### 회귀 (Regression)
- 여러 개의 독립변수와 한 개의 종속변수 간의 상관관계를 모델링하는 기법
    - 즉, 한 종속변수가 하나 이상의 독립변수에 의해 어떠한 영향을 받고   
    - 또 어떠한 관계로 나타나는지 분석하는 기법 
    - 즉, 어떠한 통계적 관계식으로 나타나는지를 밝히는 것이 목적    
- **독립 변수(X)** : 서로 간 상관관계가 적고 독립적인 변수
    - 피처 : 아파트 크기, 방 개수, 주변 지하철역 수, 주변 학군
    - 영향을 주는 변수    
- **종속 변수(Y)**
    - 결정값 : 아파트 가격
    - 영향을 받는 변수   
- **회기 계수($ w $)** : 독립 변수의 값에 영향을 미치는 것  


### 회귀식
- 선형 회귀식
     $ y = w_0 + w_1 X_{1} + w_2 X_{2} + w_3 X_{3} + ... + w_n X_{n} $

* y = 종속변수(아파트 가격)  
* X = 독립변수(방 개수, 면적, 거리 등)  
* $ w $ = 독립변수의 값에 영향을 미치는 회귀 계수 (Regression coefficients)  

### 선형 회귀
- 가장 많이 사용되는 회귀
- **실제값과 예측값의 차이(오류의 제곱값)를 최소화하는 직선**형 회귀선을 최적화하는 방식
- 실제값과 예측값의 차이인 오류를 최소로 줄일 수 있는 선형 함수를 찾아서
- 이 선형 함수에 독립변수(피처)를 입력해
- 종속변수(타깃값, 예측값)을 예측하는 것
- 규제 방법에 따라 여러 유형으로 나뉨

### 규제
- 일반적인 선형 회기의 과적합 문제를 해결하기 위해서
- 회귀 계수에 패널티 값을 적용하는 것
---
L2 규제  
- 회귀 계수 값의 제곱에 대해 패널티를 부여하는 방식  
- 회귀 계수의 크기 감소  
    
L1 규제  
- 회귀 계수 값의 절대값에 대해 패널티를 부여하는 방식  
- 영향력이 크지 않은 회귀 계수를 0으로 변환하고 제거  
- 적절한 피처만 회귀에 포함시키는 피처 선택의 특성  

### 대표적인 선형 회귀 모델
- 일반 선형 회귀 : 규제를 적용하지 않은 모델로  
    - 예측값과 실제값의 손실함수(RSS)를 최소화할 수 있도록 최적화
- 릿지(Ridge) : 선형 회귀에 L2규제를 추가한 회귀 모델
- 라쏘(Lasso) : 선형 회귀에 L1규제를 적용한 모델
- 엘라스틱넷(ElasticNet) : L2, L1 규제를 함께 결합한 모델
- 로지스틱 회귀(Logistic Regression) : 이름이 회귀일 뿐 분류에서 사용되는 선형 모델
    - 회귀의 특성을 가지고 있으면서 분류 작업 수행

### 단순 선형 회귀를 통한 회귀 이해

단순 선형 회귀  
- 독립 변수 1개, 종속 변수 1개인 선형 회귀 
- 독립변수(피처)가 하나뿐인 단순선형회귀 예
    - 주택가격이 단순히 주택의 크기로만 결정된다는 단순 선형 회귀로 가정
    - 주택가격은 주택크기에 대해 선형(직선 형태)의 관계로 표현
    - 그래프로 표시
        - X 축 : 주택 크기 (평수)
        - Y 축 : 주택 가격
    - 주택 가격 모델링
        - $ y = w_0 + w_1 * X $  
        - 이때 회귀선의 기울기 $ w_1 $이 회귀 계수가 됨   
    - 실제 가격 : $ w_0 + w_1 * X + 오류값 $ 
        - 주택의 실제 가격은 모델링한 함수에서 오류값을 뺀(또는 더한) 값  
    - 실제값 - 오류값 = 잔차
    - 최적의 회귀 모델을 만든다는 것은
        - 바로 전체 데이터의 잔차(오류값)의 합이 최소가 되는 모델을 만든다는 의미  

선형 회기 예
  



### 최적의 회귀 모델
- 실제값과 모델 사이의 오류값(잔차) 합이 최소가 되는 모델
- 오류값 합이 최소가 될 수 있는 최적의 회귀 계수를 찾는 것

### 오류의 합 구하는 방식
- 오류값은 마이너스(-) 오류와 플러스(+) 오류가 있어서 그냥 더하면 상쇄되어 원치 않는 결과가 나오므로
- 보통 오류 합을 계산할 때는 **절대값**을 취하거나 오류값의 **제곱**을 구해서 더하는 방식(RSS) 사용
- 학습을 통해 이 RSS를 최소로 하는 회귀 계수를 찾는 것이 머신러닝 기반 회귀 핵심 사항
- RSS: Residual Sum of Square (잔차제곱합) : 예측된 값과 실제 값의 차이(Error)의 제곱합  


### 머신러닝 회귀 알고리즘
- 데이터를 계속 학습하면서
- 이 오류값을 지속적으로 감소시키고
- 최종적으로 더 이상 감소하지 않는 최소 오류값을 구하는 것

### 머신러닝의 회귀 기법
- 독립변수 : Feature 
- 종속변수: 결정 값 (Target)
- 머신러닝 회귀 예측  핵심
    - 주어진 피처와 결정 값 데이터로 학습을 통해  
      - **RSS(오류값의 제곱)를 최소로 하는 최적의 회귀 계수를 찾아 예측하는 것**
    - 머신러닝에서는 손실함수의 최소값을 찾는 방식으로 경사하강법 사용
      - 손실함수(Loss Function) : 예측값과 실제값의 차이를 비교하는 함수
        - MAE, MSE, RMSE 등 
      - 경사하강법(Gradient Descent)
        - 함수의 값이 작아지는 방향으로 독립변수의 값을 변경하면서 
        - 최종 최소값을 갖게하는 독립 변수 값을 찾는 방식  
        - 랜덤 값으로 시작해서 조금씩 함수의 값이 감소하는 방향으로 
          - 최소값에 가까워질때까지 점진적으로 진행  
- 머신러닝 회귀 예측 방법    
  - (1) 주어진 피처와 결정값 데이터로 학습을 통해  
  - (2) 최적의 회귀 계수(W)를 찾아 회귀식을 만들어  
  - (3) 실제 테스트 데이터 세트에서 예측  
---
- **회귀에서 가장 중요한 것은 회귀 계수**

### 회귀 평가 지표
- MSE (Mean Squared Error)
- RMSE(Root Mean Squared Error)  
- $R^2$ : R제곱

MSE (Mean Squared Error : 평균 제곱 오차) : **실제값과 예측값의 차이를 제곱해서 평균한 것** 
- 예측값 - 관측값(데이터값)의 **제곱된 값의 평균** 
- 잔차의 제곱을 하는 이유 
    - 음수 및 오차의 합이 0이 되는 것 방지  
    - 오차의 민감도를 높이기 위함  
        - 제곱을 하기 때문에 실제 오류 평균보다 더 커짐
        - 따라서, 이상치(Outlier)에 민감  


RMSE(Root Mean Squared Error)  
- **MSE에 루트를 씌운 것**
    - MSE 값이 오류의 제곱을 구하기 때문에 실제 오류 평균보다 더 커지는 특성이 있으므로
    - MSE에 제곱근을 씌워서 계산하는 함수를 직접 만들어서 사용 
    - 특이값의 영향을 덜 받음 
        - **오차값이 크게 나오는 경우 값이 왜곡될 수 있다는 점을 보완**  
- **값이 작을수록 성능이 좋음**
- 장점 : 직관적  
- 단점  
    - 제곱하기 때문에 1미만의 에러는 작아지고, 그 이상의 에러는 커짐  
    - 실제 정답보다 낮게 예측했는지, 높게 했는지를 파악하기 힘듦  

$R^2$(R제곱) : **분산 기반으로 예측 성능 평가**
- 실제값의 분산 대비 예측값의 분산 비율을 지표로 하며
- **1에 가까울수록 예측 정확도가 높음**
