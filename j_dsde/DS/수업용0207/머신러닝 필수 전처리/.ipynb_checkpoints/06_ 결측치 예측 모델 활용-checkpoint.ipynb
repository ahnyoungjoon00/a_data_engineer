```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt #   
%matplotlib inline
```

```python
df = pd.read_csv(".//mammographic.csv")
```

```python
df.info()
df.head()
```

```python
X = df.drop("Output", axis=1)
y= df["Output"]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=123)
```

```python
X_train.isna().sum() / len(X_train)
```

```python
X_train.corr().sum()/len(X_train.columns)
```

```python
from sklearn.impute import KNNImputer
# 1.   
ki = KNNImputer(n_neighbors=5)
# 2. fit(   )
ki.fit(X_train)
# 3. transform(     )
ki.transform(X_train)
```

```python
X_train = pd.DataFrame(ki.transform(X_train), columns=X_train.columns)
X_train.isna().sum()

#  fit X_train     
X_test = pd.DataFrame(ki.transform(X_test), columns=X_test.columns)
X_test.isna().sum()
```
