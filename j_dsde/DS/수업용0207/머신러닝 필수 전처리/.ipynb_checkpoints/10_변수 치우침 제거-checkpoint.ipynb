- 모델링에 가장 적합한 확률분포는 정규분포이나 실제로 많은 변수가 특정방향으로 치우쳐 있음
- 한쪽으로 치우친 변수에서 치우친 반대 방향의 값(꼬리부분)들이 이상치처럼 작용할 수 있으므로, 이러한 치우침을 제거해야 함

### 탐색방법 : 왜도(skewness)

- 변수 치우침을 확인하기 가장 적절한 척도로는 왜도가 있음
- 왜도
    - 분포의 비대칭도를 나타내는 통계량
    
    
    - 왜도의 절대값이 <b> 1.5이상 </b> 이면 치우쳤다고 판단

```python
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"

import warnings
warnings.filterwarnings(action='ignore')

import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("./데이터/sonar.csv", header=None)
df.info()
df.head()
```

```python
len(df.columns)
```

```python
# 컬럼명 생성
columns = ["band"+str(i) for i in range(1, len(df.columns))]
columns.append("Y")
df.columns = columns
```

```python
df.head()
```

```python
X = df.drop("Y", axis= 1)
y = df["Y"]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=123)
X_train.shape
X_test.shape
```

#### 치우침 제거

```python
# 변수 치우침 정도 확인 : skew()
# 양의 왜도 : 왼쪽으로 치우침, 오른쪽으로 꼬리가 긴, 최빈값 < 중위값 < 평균값
# 0
# 음의 왜도 : 오른쪽으로 치우침, 왼쪽으로 꼬리가 긴, 평균값 < 중위값 < 최빈값
X_train.skew()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
print("최빈값 : ", df["band4"].mode(), "\n",
      "중위값 : ", df["band4"].median(), "\n",  
      "평균값 : ", df["band4"].mean(), "\n",)
df["band4"].hist()
```

```python
# 종속변수 문자 >> 숫자로 변환 필요
# pd.set_option("future.no_silent_downcasting", True) # 이거 옛날함수라고 경고뜨는거 없애는
y_train.replace({"M":-1, "R":1}, inplace=True)
y_test.replace({"M":-1, "R":1}, inplace=True)
```

#### 치우침 제거 : 로그 변환
- 학습을 위해 기존 값을 다른 값으로 바꾸는 것은 가능(모든 데이터가 동일한 변환을 진행하게 되면 데이터의 성질은 그대로 유지된다
- skew의 절대값이 1.5 이상이면 치우쳐져 있다고 결정

```python
# skew절대값이 1.5이상인 컬럼만 추출
biasedVar = X_train.columns[X_train.skew().abs() >=1.5]
biasedVar
```

```python
X_train[biasedVar].head()
```

```python
X_train[biasedVar]["band4"].hist()
plt.show()
```

```python
np.log(X_train[biasedVar]["band4"]).hist()
plt.show()
```

#### 왜도가 1.5이상인 컬럼 모두 log 변환

```python
X_train[biasedVar] = np.log(X_train[biasedVar])
```

```python
X_train.skew().abs() >=1.5
```

### X_test에 적용
- log 변환 이후에 학습을 시킨 모델로 실제값을 예측하기때문에
- log 변환된 학습데이터의 컬럼은 test데이터에서도 동일하게 변환되어야한다.

```python
X_test[biasedVar] = np.log(X_test[biasedVar])
```
