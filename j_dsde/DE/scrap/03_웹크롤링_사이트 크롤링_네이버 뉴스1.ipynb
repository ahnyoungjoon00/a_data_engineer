### 네이버 뉴스 섹션 메뉴와 섹션별 URL 추출

```python
# 사이트 연결. HTML 파싱
url = 'https://news.naver.com/'
html = urlopen(url)
# 파서 객체 생성
bs_obj = BeautifulSoup(html, 'html.parser') 
```

```python
bs_obj
```

```python
# bs_obj 내용 확인했을 때 웹 페이지의 텍스트 내용이 출력되지 않고 
# <Script> 태그만 있으면 BeautifulSoup으로 파싱 불가 
# Selenium 사용해서 동적 크롤링 해와야 함
```

```python
# 메뉴 항목 추출 : 언론사별 정치 경제 ...
#  메뉴 ul 태그를 사용하므로 ul기준 확인 : F12로, 타겟컨텐츠 확인
#  -> ul태그, class:Nlnb_menu_list

# finde 함수로 뉴스분류 메뉴탭탭
ul = bs_obj.find("ul", {"class":"Nlnb_menu_list"})
lis=ul.find_all("li")
lis

bs_obj.find("ul", {"class":"Nlnb_menu_list"}).find_all("li")
```

```python
# select함수 사용
ul =bs_obj.select_one(".Nlnb_menu_list")
lis = ul.select("li")
lis
len(lis)
# bs_obj.select_one(".Nlnb_menu_list").select("li")
# bs_obj.select(".Nlnb_menu_list > li")
```

```python
lis[0]
lis[0].a # a태그를 이런식으로 불러버려도 됨
lis[0].span
lis[0].text
lis[0].a["href"]
```

```python
# 각 메뉴의 항목과 링크를 반복문으로 추출
for li in lis:
    print(li.text, li.a["href"])
```

```python
# 수집한 데이터를 csv 파일로 저장
# (1) 섹션 메뉴와 링크를 리스트로 생성
# (2) 데이터프레임 생성
# (3) 파일로 저장
```

```python
# (1) 섹션 메뉴와 링크를 리스트로 생성

section=[]
link=[]

lis = bs_obj.select("ul.Nlnb_menu_list > li")

for li in lis:
    section.append(li.text)
    link.append(li.a["href"])
# section
# link
```

```python
# (2) 데이터프레임 생성
# pandas 라이브러리 사용
# pd.DataFrame({'칼럼명':리스트, '칼러명':리스트})

len(section), len(link)
df = pd.DataFrame({"menu":section, "link":link})
```

```python
# (3) 파일로 저장
# source_code 폴더 안에 crawl_data 폴더 생성
# 데이터프레임.to_csv('파일경로 및 파일명', index=0)
# index=0 : index는 저장하지 않음

# df.to_csv("./crawl_data/new_section_df.csv", index=0)
df.to_csv("./crawl_data/new_section_df.csv", index=0, encoding="utf-8-sig")
```
