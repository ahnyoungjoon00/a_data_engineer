## 공공데이터 API 사용

### 금융위원회_KRX상장종목정보
- 한국 거래소에서 제공하는 KRX에 상장된 종목에 대한 정보  
- KRX : Korea Exchange (한국 거래소) 

### 활용 신청
- 데이터 찾기 > 데이터목록
- 금융위원회_KRX상장종목정보  검색
- 오픈 API(건) 에서 XML 형식의 데이터 : [활용신청]


#### 활용 신청 시 선택 및 입력 사항  
- 활용 목적 : 참고자료 선택  
- 입력 : 데이터 분석 참고자료  
- 동의합니다 : 체크  

#### [마이페이지]에서 확인
- [마이페이지] 오른쪽 목록에서 
	- [승인] 금융위원회_KRX상장종목정보  클릭하면 [개발계정 상세보기]로 이동  
	- 서비스정보에서 [일반 인증키] 복사해서 잘 보관   
             - [활용신청 상세기능정보] 에서 미리보기 [확인]  
		- serviceKey : 본인 키 입력  
		- numOfRows : 10  
		- beginBasDt : 20240101 (이상 검색)  
		- [미리보기] 하면 결과 출력
