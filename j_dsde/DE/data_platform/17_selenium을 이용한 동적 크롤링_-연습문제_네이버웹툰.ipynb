```python
def get_page_webtoon(url):
    try :
        driver.get(url)
        time.sleep(1)
        # 함수로 작성하기 위해 정리
        titles = driver.find_elements(By.CSS_SELECTOR, 'div.ChallengeListItem__info_area--hnJTz a.ContentTitle__title_area--x24vt')
        authors = driver.find_elements(By.CSS_SELECTOR, 'div.ChallengeListItem__info_area--hnJTz a.ContentAuthor__author--CTAAP')
        ratings = driver.find_elements(By.CSS_SELECTOR, 'div.ChallengeListItem__rating_area--rU1Lc span.Rating__star_area--dFzsb .text')
        views = driver.find_elements(By.CSS_SELECTOR, 'div.ChallengeListItem__rating_area--rU1Lc span.Rating__view_area--GQb_S .text')
        
        title_list = [ title.text.replace('UP\n','')  for title in titles]
        author_list = [ author.text for author in authors]
        rating_list = [ rating.text for rating in ratings]
        view_list = [ view.text for view in views]

    # 한 페이지를 읽어 온 뒤 DF로 구성
        # {1} 빈 데이터프레임 생성
        naver_webtoon_df = pd.DataFrame({
            "제목":title_list,
            "작가":author_list,
            "별점":rating_list,
            "조회수":view_list,
        })
        return naver_webtoon_df
        
    except:
        return False
```

```python
url ='https://comic.naver.com/bestChallenge?page=1'
get_page_webtoon(url)
```

```python
# {1} 빈 데이터프레임 생성
naver_webtoon_finalDF = pd.DataFrame({
    "제목":[],
    "작가":[],
    "별점":[],
    "조회수":[],
})
```

```python
base_url ='https://comic.naver.com/bestChallenge?page='

for i in range(1, 11):
    print('%2d 페이지 추출중' % i)
    url = base_url+str(i)
    df = get_page_webtoon(url)
    naver_webtoon_finalDF = pd.concat([naver_webtoon_finalDF, df], axis = 0, ignore_index=True)
naver_webtoon_finalDF
```

```python
naver_webtoon_finalDF.to_csv("../scrap/crawl_data/naver_webtoon.csv", index = 0, encoding = "utf-8-sig")
```
